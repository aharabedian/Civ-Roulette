```python
from bs4 import BeautifulSoup
from pandas import DataFrame
```

# Scrape CIV V Base Civs

```python
html = open("civ_v.html", "r")
soup = BeautifulSoup(html, 'html')
```

```python
headers = [data.text for data in soup.find("thead").find("tr").find_all("td")]
headers
```

```python
def process_civ_data(civ_data):
    raw_data = [data.text for data in civ_data.find_all("td")]
    data = {}
    for i in range(7):
        data[headers[i]] = raw_data[i]
    
    data["DLC"] = ""

    # Extract DLC
    if data["Civilization"][-5:] == "(DLC)":
        data["Civilization"] = data["Civilization"][:-5].strip()
        data["DLC"] = data["Civilization"] + " DLC"
    
    if data["Civilization"][-5:] == "(G&K)":
        data["DLC"] = "Gods & Kings"
        data["Civilization"] = data["Civilization"][:-5].strip()
    
    if data["Civilization"][-5:] == "(BNW)":
        data["DLC"] = "Brave New World"
        data["Civilization"] = data["Civilization"][:-5].strip()
    
    return data

complete_data = [process_civ_data(row) for row in soup.find("tbody").find_all("tr")]
complete_data
```

```python
civ_data = DataFrame(complete_data)
civ_data
```

```python
civ_data.to_csv("civ_v_base.csv", index=False)
```
